```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/WinifredIfy/VBreakdown/main/VBreakdown.csv")
df.head()
```

```python
df.head(100)
```

```python
# CLEANING DATA
```

```python
# Drop rows with any missing values
```

```python
df.dropna(inplace=True)
```

```python
# Drop Duplicates
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.rename(columns={'VEHICLES': 'Vehicles', 'VEH_LAST_MILEAGE': 'Vehicle_last_mileage','VEH_AGE_IN_MONTHS': 'Age', '3MONTHS_BREAKDOWN_RECORD': 'Breakdowns'}, inplace=True)
```

```python
# ANALYSISING AND VISUALIZING OF DATA BY PLOTTING BAR CHARTS
```

```python
#Plot chart of breakdowns against the age of each vehicle
Age = df['Age']
Breakdowns = df['Breakdowns']

plt.bar(Age, Breakdowns,width=10, color="#370726")

plt.xlabel("age in months")
plt.ylabel("breakdowns")
plt.title("chart of breakdowns against the age (months) of each vehicle")

plt.legend()
plt.grid(False)
plt.show()
```

```python
#Plot chart of breakdowns against the mileage of each vehicle
Breakdowns = df['Breakdowns']
Mileage = df['Vehicle_last_mileage']

plt.bar(Breakdowns, Mileage,  color="#370726")

plt.xlabel("breakdowns")
plt.ylabel("mileage")
plt.title("chart of breakdowns against the mileage of each vehicle")

plt.legend()
plt.grid(False)
plt.show()
```

```python
#Plot a graph of age and mileage to determine a correlation between the two variables
Age = df['Age']
Mileage = df['Vehicle_last_mileage']

plt.bar(Age, Mileage,width=10, color="#370726")


plt.xlabel("Age in months")
plt.ylabel("Mileage")
plt.title("chart of Age (months) against the mileage of each vehicle")

plt.legend()
plt.grid(False)
plt.show()
```
